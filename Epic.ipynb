# Epic
## What is Epic?
---


Epic is an interactive horror fiction writer. By selecting from a set of authors, subgenres, and tuning your generating parameters, Epic will generate a story for you.

## Instructions:
---


1. All you have to do to install and run Epic is run the cell below. ***Note that this process may take awhile.***
2. Once installation has finished, a few links will be outputted at the bottom. The link to the application is the one containing "ngrok".
3. Welcome to Epic! Start by selecting some authors you would like to have you story styled like.
4. By clicking on the genre tab at the top, you will be able to choose your subgenres.
5. Finally, the last time will help you control how Epic generates your story. The Blacklist words are words that will not appear in your generated text.
6. Clicking the 'Get Writing' button will save your settings and start the story.
7. You can enter text either in the main text area or in the input box. Don't worry if you don't provide any input as the writer will start off for you. Don't forget to set a title!
8. Clicking the 'Continue' button will start the Epic text generation process. Text will appear above when the model is done generating. If you don't like the output, simply change or delete it and continue from there.
9. If you want to change your settings, simply click on the hamburger menu at the top and set up the parameters you want. **DON'T FORGET TO CLICK 'GET WRITING!' TO SAVE YOUR SETTINGS**. If you don't want to save your settings, simply click the 'X' to return to your writing.
10. Happy Writing!

## Notes on Generation
---


* Using a longer Generation Length will increase the time it takes for the writer to generate.
* Giving a good start to the story helps create a good story. Not providing a start will work, but your results may not be as good.
* Epic makes writing errors sometimes, but the output is editable.
* The model has been trained to identify specific literary elements from context such as plot, climax, action, resolution, and many more. By giving the model a little direction towards ones of elements, Epic will pick up on this and generate text to match.
* If Epic isn't writing how you like, or the output is not very legible, try adding more authors. Authors like Stephen King, Bram Stoker, and other popular horror novelist make up a larger part of the dataset and generally provide the best outputs.
